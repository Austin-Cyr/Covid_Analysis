```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
county_deaths = pd.read_csv("../Death_By_County/Death_By_County(HL).csv")
county_deaths.head()
```

```python
county_vax = pd.read_csv("../Vaccine_By_County/Formatted_Vax_by_County.csv")
county_vax.head()
```

```python
ml_df = pd.merge(county_vax, county_deaths ,on='FIPS',how='outer')
ml_df
```

```python
ml_df = ml_df.drop("Recip_County", 1)
ml_df = ml_df.drop("Recip_State", 1)
# ml_df = ml_df.drop("State", 1)
# ml_df = ml_df.drop("County Name", 1)
```

```python
ml_df = ml_df.rename({"Series_Complete_Pop_Pct": "Fully_Vaccinated_%"}, axis=1)
ml_df
```

```python
pop_data = pd.read_csv("../Austin_population_Data.csv")
pop_data.head()
```

```python
ml_df2 = pd.merge(ml_df, pop_data ,on='FIPS',how='outer')
ml_df2
```

```python
ml_df2 = ml_df2.dropna()
ml_df2
```

```python
machine_learning_df = ml_df2.filter(["FIPS", "State", "County Name", "Urban Rural Code", "Urban Rural Description", "POPESTIMATE2020", "Total_Vaccines", "% Vaccinated", "Case_Count", "% Pop Cases", "Total deaths", "COVID-19 Deaths", "Pct_Deaths_by_Covid", "Covid_Death_Pct_Above_Median"], axis=1)
machine_learning_df
```

```python
machine_learning_df = machine_learning_df.rename({"POPESTIMATE2020": "Population Estimate 2020", "% Vaccinated": "Percent Vaccinated" , "% Pop Cases": "Percent of Population Cases", "Total_Vaccines": "Total Vaccines", "Case_Count": "Case Count", "Pct_Deaths_by_Covid": "Percent Deaths by Covid", "Covid_Death_Pct_Above_Median": "Covid Death Percent Above Median"}, axis=1)
```

```python
machine_learning_df["Percent Vaccinated"] = machine_learning_df["Percent Vaccinated"] * 100
machine_learning_df["Percent of Population Cases"] = machine_learning_df["Percent of Population Cases"] * 100
```

```python
machine_learning_df
```

```python
machine_learning_df.to_csv('Machine_Learning_Data.csv', index= False)
```

## CREATE A DF FOR POPULATION DEATHS BY COVID

```python
death_by_pop_df = machine_learning_df
```

```python
death_by_pop_df["Percent_Pop_Death_by_Covid"] = death_by_pop_df["COVID-19 Deaths"]/ death_by_pop_df["Population Estimate 2020"]*100
death_by_pop_df.head()
```

```python
death_by_pop_df["Percent_Pop_Death_by_Covid"].median()
```

```python
def create(row):
  if row["Percent_Pop_Death_by_Covid"] > .24:
    return 1
  else:
    return 0
death_by_pop_df["Pct_Pop_Died_by_Covid_Above_Median"] = death_by_pop_df.apply(create, axis=1)
```

```python
death_by_pop_df.head()
```

```python
death_by_pop_df = death_by_pop_df.filter(["FIPS", "State", "County Name", "Urban Rural Code", "Urban Rural Description", "Population Estimate 2020", "Total Vaccines", "Percent Vaccinated", "Case Count", "Percent of Population Cases", "COVID-19 Deaths", "Percent_Pop_Death_by_Covid", "Pct_Pop_Died_by_Covid_Above_Median"])
death_by_pop_df.head()
```

```python
death_by_pop_df.to_csv('Pct_Death_by_Covid.csv', index= False)
```
