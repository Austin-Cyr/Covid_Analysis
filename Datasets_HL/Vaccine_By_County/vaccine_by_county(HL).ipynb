```python
import pandas as pd
```

```python
vax_by_cty_df = pd.read_csv("US_Vaccine_By_County.csv")
```

```python
vax_by_cty_df.head(50)
```

```python
vax_by_cty_condensed = vax_by_cty_df.filter(["Date", "FIPS","Recip_County","Recip_State", "Series_Complete_Pop_Pct","Metro_status"], axis=1)
```

```python
vax_by_cty_condensed.head()
```

```python
vbc_df = vax_by_cty_condensed.dropna()
```

```python
vbc_df
```

```python
vbc_latest_df = vbc_df.drop_duplicates(subset=['FIPS'])
```

```python
vbc_latest_df.to_csv('Formatted_Vax_by_County.csv')
```
