```python
import pandas as pd
```

```python
health_data = pd.read_csv("../Obesity_and_Health/analytic_data2021.csv")
health_data
```

```python
health_data
```

```python
#Info From University of Wisconson
health_df = health_data.filter(["5-digit FIPS Code", "Adult obesity raw value", "Adult smoking raw value", "Poor of fair health raw value", "Excessive drinking raw value"])
health_df
```

```python
health_df = health_df.rename({"5-digit FIPS Code": "FIPS", "Adult obesity raw value":"Adult Obesity Percent", "Adult smoking raw value": "Adult Smoking Percentage", "Excessive drinking raw value": "Excessive Drinking Percent"}, axis=1)
```

```python
health_df = health_df[1:]
health_df
```

```python
health_df.dtypes
```

```python
health_df["FIPS"]= health_df["FIPS"].astype(str).astype(int)
health_df["Adult Obesity Percent"]= health_df["Adult Obesity Percent"].astype(str).astype(float)
health_df["Adult Smoking Percentage"]= health_df["Adult Smoking Percentage"].astype(str).astype(float)
health_df["Excessive Drinking Percent"]= health_df["Excessive Drinking Percent"].astype(str).astype(float)
```

```python
health_df.to_csv('County_Health_Clean.csv', index=False)
```
