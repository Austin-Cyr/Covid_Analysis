```python
import pandas as pd
```

```python
age_df = pd.read_csv("output.csv")
age_df
```

```python
#Get entries from latest year only
age_df = age_df[age_df['YEAR'] == 12]
age_df.head()
```

```python
#Drop year column
age_df = age_df.drop('YEAR', axis=1)
age_df.head()
```

```python
#Check unique state names to be sure all are here
# n = len(pd.unique(age_df["STNAME"]))
# n
pd.unique(age_df["STNAME"])
```

```python
# add leading zeros to "County" column to create FIPS name
age_df["COUNTY"] = age_df['COUNTY'].apply(lambda x: '{0:0>3}'.format(x))
age_df.head()
```

```python
#Join state and county to create fips code
age_df["FIPS"] = age_df["STATE"].astype(str) + age_df["COUNTY"].astype(str)
age_df.head()
```

```python
age_by_county = age_df.filter(["FIPS", "STNAME", "CTYNAME", "MEDIAN_AGE_TOT"])
age_by_county
```

```python
age_by_county.to_csv('Median_Age_County.csv', index= False)
```
