```python
import pandas as pd
import numpy as np
```

```python
death_by_county_df = pd.read_csv("Provisional_COVID-19_Deaths_by_County__and_Race_and_Hispanic_Origin.csv")
```

```python
death_by_county_df.head()
```

```python
death_by_county_condensed = death_by_county_df.filter(["Data as of", "State","County Name","Urban Rural Code", "FIPS Code", "Indicator", "Total deaths", "COVID-19 Deaths", "Urban Rural Description",], axis=1)
```

```python
death_by_county_condensed.head()
```

```python
death_by_county_condensed["Pct_Deaths_by_Covid"] = death_by_county_condensed["COVID-19 Deaths"] / death_by_county_condensed["Total deaths"] * 100
```

```python
death_by_county_condensed
```

```python
death_by_county_condensed["Pct_Deaths_by_Covid"].describe()
```

```python
death_by_county_condensed["Pct_Deaths_by_Covid"].median()
```

```python
def create(row):
  if row['Pct_Deaths_by_Covid'] > 12.07:
    return 1
  else:
    return 0
death_by_county_condensed["Covid_Death_Pct_Above_Median"] = death_by_county_condensed.apply(create, axis=1)
```

```python
death_by_county_condensed
```

```python
death_by_county_condensed = death_by_county_condensed.rename({"FIPS Code": "FIPS"}, axis=1)
death_by_county_condensed
```

```python
deaths_df = death_by_county_condensed[death_by_county_condensed["Indicator"] == "Distribution of COVID-19 deaths (%)"]
deaths_df
```

```python
deaths_df = deaths_df.filter(["State","County Name","Urban Rural Code", "FIPS", "Total deaths", "COVID-19 Deaths", "Urban Rural Description","Pct_Deaths_by_Covid", "Covid_Death_Pct_Above_Median"], axis=1)
```

```python
deaths_df.to_csv('Death_by_County(HL).csv', index=False)
```
