## Database Viewing

```python
# SQLAlchemy
from sqlalchemy import create_engine
from sqlalchemy import inspect
# Pandas
import pandas as pd

# Path to sqlite
database_path = "covid_2.db"
```

```python
# Create an engine that can talk to the database
engine = create_engine(f"sqlite:///{database_path}")
```

```python
# Get the name of the table. 
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# Using the inspector to print the column names within the 'Census_Data' table and its types
columns = inspector.get_columns('health_data')
pd.DataFrame(columns)
```

```python
# Using the inspector to print the column names within the 'Census_Data' table and its types
columns = inspector.get_columns('machine_learning_data')
pd.DataFrame(columns)
```
